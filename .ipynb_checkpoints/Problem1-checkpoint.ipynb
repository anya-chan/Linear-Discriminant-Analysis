```python
from GaussianDiscriminantAnalysis import GaussianDiscriminantAnalysis
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1 = np.array([[2, 3, 4, 3, 5, 5, 2, 4, 0, 6],[3, 2, 4, 2, 4, 3, 8, 6, 0, 5]]).T
X2 = np.array([[7, 9, 10, 11, 9, 13, 9, 11, 11, 10],[16, 15, 13, 11, 17, 13, 14, 15, 14, 12]]).T

#x1 = np.array([[4, 2, 2, 3, 4],[1, 4, 3, 6, 4]]).T
#x2 = np.array([[9, 6, 9, 8, 10],[10, 8, 5, 7, 8]]).T

[w,w0] = GaussianDiscriminantAnalysis(X1, X2)

xtest = np.array([[4, 10, 10],[0, 2.5, 12.5]])

a = w@xtest+w0
y = 0

print("a = ")
print(a)

for i in a:
    y = y+1
    if i >=0:
        print("The " + str(y) + " datapoint is class 1")
    else:
        print("The " + str(y) + " datapoint is class 2")
        
#y_pred1_train = gda1.predict(X_train1)
#print(y_pred1_train)
```
